```
import pandas as pd

# ubah dataset jadi dataframe
df = pd.read_csv('diabetes.csv')
# lihat lima data pertama
# df.head()

# mengecek apakah terdapat nilai-nilai yang hilang pada dataset serta apakah ada atribut yang bukan berisi bilangan numerik
# df.info()

# memisahkan atribut pada dataset dan menyimpannya pada sebuah variabel
X = df[df.columns[:8]]
# memisahkan label pada dataset dan menyimpannya pada sebuah variabel
y = df['Outcome']

from sklearn.preprocessing import StandardScaler
# standarisasi skala nilai-nilai dari dataset
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

from sklearn.model_selection import train_test_split
# memisahkan data untuk training dan testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)

from sklearn.svm import SVC
 
#  membuat objek Support Vector Classifier
clf = SVC()
# memanggil fungsi fit untuk melatih model
clf.fit(X_train, y_train)

# melihat akurasi prediksi dari model yang kita latih terhadap data testing
clf.score(X_test, y_test)
```
